```python
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

###### Barplot

```python
sns.barplot(x="sex", y="total_bill", data=tips)
```

```python
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.std) # MÃ©dia da conta
```

##### Countplot

```python
sns.countplot(x="sex", data=tips) # Numero de homens e de mulheres
```

##### Boxplot

```python
sns.boxplot(x="day", y="total_bill", data=tips) 
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="rainbow", hue="sex") 
```

```python
sns.boxplot(data=tips, orient="h")
```

##### Violinplot

```python
sns.violinplot(x="day", y="tip", data=tips)
```

```python
sns.violinplot(x="day", y="tip", data=tips, hue="sex")
```

```python
sns.violinplot(x="day", y="tip", data=tips, hue="sex", split=True)
```

##### Stripplot

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=False)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True)
```

##### Swarnplot

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, color="black")
sns.violinplot(x="day", y="total_bill", data=tips)
```

##### Factorplot

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="bar")
```
