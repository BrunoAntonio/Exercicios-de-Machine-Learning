```python
import matplotlib.pyplot as plt
```

```python
# para evitar escrever plt.show() para exibir os gráficos
%matplotlib inline
```

```python
import numpy as np
```

```python
x =np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

##### Funcional

```python
plt.plot(x, y, color ="r")
plt.xlabel("Eixo x")
plt.ylabel(" Eixo y")
plt.title("Título")
```

```python
plt.plot(x, y, "r--")
```

##### Subplot

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r--")

plt.subplot(1, 2, 2)
plt.plot(y, x, "g*-")
```

##### Figure

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.3, 0.3, 0.3])

axes1.plot(x, y)
axes1.set_xlabel("Eixo x")
axes1.set_title("Título")

axes2.plot(y, x)
```

##### Subplots e Figure

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, "b--")
ax.plot(x, x**4, "g-.")

ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
ax.set_title("Título")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
for axe in ax:
    axe.plot(x, y, "b")
    axe.set_title("Título")
fig
```

Outro modo de preencher os gráficos

```python
ax[0].plot(x, y, "b")
ax[0].set_title("Título")

ax[1].plot(x, y, "r")
ax[1].set_title("Título")

fig
```

tight_layout() para não haver sobrepossições

```python
fig, ax = plt.subplots(5, 5)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, "r-", label="x^2")
axes.plot(y, x, "g--", label="x**1/2")

axes.set_title("Título")
axes.legend(loc=1) #loc para definir a possição da legenda
```

Gravar o gráfico

```python
fig.savefig("imagem.jpg")
```

##### Modificar a vizualização dos gráficos

```python
x
```

```python
fig, ax =plt.subplots(figsize=(8,10))
ax.plot(x, x**2, color="green", linewidth=10, alpha=0.2, linestyle="--")
ax.plot(x, x**3, color="blue", linestyle=":", marker="o", markersize=8, markerfacecolor="red")
ax.plot(x, x**4, color="yellow", marker="o", markersize=8, markerfacecolor="red", markeredgewidth=3, markeredgecolor="blue")


#Definir limites do Eixo X e Y
ax.set_xlim(0,2)
ax.set_ylim(0,10)


# Possiveis estilos de linha: ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
# possíveis símbolos de marcador: marcador = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
```
